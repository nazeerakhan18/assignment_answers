### Actual Output

```python

```

Python uses **short-circuit evaluation**, which means it stops evaluating an expression as soon as the final result is determined.

- `0 or 5` → `5` (0 is False, so it returns the next value.)
- `5 or 0` → `5` (5 is True, so it stops immediately.)
- `None or "hello"` → `"hello"` (`None` is False.)
- `0 and 5` → `0` (0 is False, so it stops immediately.)
- `5 and 10` → `10` (Both are True, so it returns the last value.)
- `[] or "default"` → `"default"` (An empty list is False.)
- `[0] or "default"` → `[0]` (A non-empty list is True, even if it contains 0.)

````python
# Q18 - Temperature Converter


temperature = float(input("Enter the temperature: "))
unit = input("Enter the unit (C/F): ").upper()

if unit == "C":
    fahrenheit = (temperature * 9 / 5) + 32
    print(f"{temperature:.2f}°C = {fahrenheit:.2f}°F")

elif unit == "F":
    celsius = (temperature - 32) * 5 / 9
    print(f"{temperature:.2f}°F = {celsius:.2f}°C")

else:
    print("Invalid unit! Please enter C or F.")
````

This program checks whether a given number is an Armstrong number by calculating the sum of each digit raised to the power of the total number of digits.

````python
# Q24 - Armstrong Number Checker

number = int(input("Enter a number: "))

original = number
digits = len(str(number))
total = 0

while number > 0:
    digit = number % 10
    total = total + (digit ** digits)
    number = number // 10

if total == original:
    print(original, "is an Armstrong Number.")
else:
    print(original, "is not an Armstrong Number.")
````

Find all Armstrong numbers between 1 and 9999. (An n-digit number is Armstrong if the sum of
its digits each raised to the power n equals the number itself. Example: 153 = 1³ + 5³ + 3³.) Print them
in a clean comma-separated list with a count at the end.

This program checks every number from 1 to 9999 and prints all Armstrong numbers in a comma-separated list along with the total count.

````python
# Q31 - Find All Armstrong Numbers (1 to 9999)

count = 0

print("Armstrong Numbers:")

for number in range(1, 10000):

    original = number
    temp = number
    digits = len(str(number))
    total = 0

    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10

    if total == original:
        print(original, end=", ")
        count += 1

print("\n")
print("Total Armstrong Numbers =", count)
````

#Q32.
Take N as input. Print the first N terms of the Fibonacci series. Also compute and print the SUM
of those N terms. Do NOT use recursion. Edge case: N = 0 should print nothing, sum = 0. N = 1 should
print just 0.


This program prints the first N terms of the Fibonacci series without using recursion and calculates the sum of those terms.

````python
# Q32 - Fibonacci Series with Sum

n = int(input("Enter the value of N: "))

a = 0
b = 1
sum_of_terms = 0

print("Fibonacci Series:")

for i in range(n):
    print(a, end=" ")
    sum_of_terms += a

    next_term = a + b
    a = b
    b = next_term

print("\n")
print("Sum =", sum_of_terms)
````

#Q33.
 Prime check, optimized. Take a number N. Check if it is prime by iterating only up to √N. Then,
print the SUM of all primes between 2 and N (inclusive). Do not use any library.

This program prints all prime numbers from 2 to N and calculates their sum. The prime check is optimized by testing divisibility only up to the square root of the number.

````python
# Q33 - Prime Numbers up to N

n = int(input("Enter the value of N: "))

sum_of_primes = 0

print("Prime Numbers:")

for number in range(2, n + 1):

    is_prime = True

    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            is_prime = False
            break

    if is_prime:
        print(number, end=" ")
        sum_of_primes += number

print("\n")
print("Sum =", sum_of_primes)
````

#Q34.
Take a number as input. Reverse it WITHOUT converting to string. Then check if the original and
reversed are equal (palindrome). Handle negative numbers, they should be considered NOT
palindrome (because of the minus sign). Examples: 121 → palindrome, 123 → not palindrome, -121
→ not palindrome.

This program reverses a number without converting it into a string and checks whether the original number is a palindrome. Negative numbers are considered not palindrome.

````python
# Q34 - Reverse Number and Palindrome Check

number = int(input("Enter a number: "))

if number < 0:
    print("Negative numbers are not palindrome.")

else:
    original = number
    reverse = 0

    while number > 0:
        digit = number % 10
        reverse = reverse * 10 + digit
        number = number // 10

    print("Reversed Number =", reverse)

    if original == reverse:
        print("Palindrome Number")
    else:
        print("Not a Palindrome")
````

#Q35.
 Take two positive integers. Compute their HCF (Highest Common Factor) using Euclidean
algorithm with a `while` loop. Then compute LCM using the relation: HCF × LCM = a × b. Do not use
math.gcd.

This program calculates the HCF (Highest Common Factor) of two positive integers using the Euclidean algorithm with a `while` loop. It then calculates the LCM using the relation:LCM = (a × b) / HCF


````python
# Q35 - HCF and LCM using Euclidean Algorithm

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

num1 = a
num2 = b

while b != 0:
    remainder = a % b
    a = b
    b = remainder

hcf = a

lcm = (num1 * num2) // hcf

print("HCF =", hcf)
print("LCM =", lcm)
````

#Q36.
 String scanner, single pass. Take a sentence (e.g., a Swiggy review like "Order came on time,
food was hot, packaging 10/10!"). In ONE pass through the string, count: total characters, total
words, vowels (a, e, i, o, u - case-insensitive), consonants, digits, spaces, and special characters. Print
all counts

This program scans a string only once and counts the number of vowels, consonants, digits, spaces, and special characters.

````python
# Q36 - String Scanner (Single Pass)

text = input("Enter a string: ")

vowels = 0
consonants = 0
digits = 0
spaces = 0
special = 0

for ch in text:

    if ch.lower() in "aeiou":
        vowels += 1

    elif ch.isalpha():
        consonants += 1

    elif ch.isdigit():
        digits += 1

    elif ch.isspace():
        spaces += 1

    else:
        special += 1

print("\nResults:")
print("Vowels =", vowels)
print("Consonants =", consonants)
print("Digits =", digits)
print("Spaces =", spaces)
print("Special Characters =", special)
````

 (a) maximum,

 (c) average,
(d) count of students above average,
(e) the SECOND highest mark.

This program compresses a string by replacing consecutive repeated characters with the character followed by its count.

````python
# Q37 - Run-Length Encoding (RLE)

text = input("Enter a string: ")

if len(text) == 0:
    print("Empty string")

else:
    count = 1

    for i in range(1, len(text)):

        if text[i] == text[i - 1]:
            count += 1

        else:
            print(text[i - 1] + str(count), end="")
            count = 1
    print(text[-1] + str(count))
````

 Word Frequency Analyzer for Swiggy reviews. Take a review sentence. Find the frequency of
each word (case-insensitive, "Food" and "food" should be counted together). Print words sorted by
frequency in descending order. Also print the MOST frequent word and its count. Use a dict.


This program checks whether two strings are anagrams by comparing the frequency of each character. It ignores spaces and letter case without using `sorted()`.

````python
# Q38 - Anagram Checker

text1 = input("Enter first string: ")
text2 = input("Enter second string: ")

text1 = text1.replace(" ", "").lower()
text2 = text2.replace(" ", "").lower()

if len(text1) != len(text2):
    print("They are NOT Anagrams.")

else:
    count1 = {}
    count2 = {}

    for ch in text1:
        count1[ch] = count1.get(ch, 0) + 1

    for ch in text2:
        count2[ch] = count2.get(ch, 0) + 1

    if count1 == count2:
        print("They are Anagrams.")
    else:
        print("They are NOT Anagrams.")
````

TCS stock prices for a week stored as a tuple: `prices = (3845, 3902, 3878, 3955, 4012, 3998,
4055)` (Mon to Sun in ₹). Find:

(a) maximum price and which day,

(b) minimum price and which day,

(c) average,

 (d) biggest single-day gain (and between which days),

 (e) total volatility = sum of
absolute day-to-day price changes.


This program analyzes TCS stock prices stored in a tuple. It finds the maximum and minimum prices with their corresponding days, calculates the average price, identifies the biggest single-day gain, and computes the total volatility.

````python
# Q41 - TCS Stock Price Analysis

prices = (3845, 3902, 3878, 3955, 4012, 3998, 4055)
days = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday")

max_price = max(prices)
min_price = min(prices)

print("Maximum Price = ₹", max_price)
print("Day =", days[prices.index(max_price)])

print("\nMinimum Price = ₹", min_price)
print("Day =", days[prices.index(min_price)])

average = sum(prices) / len(prices)

print("\nAverage Price = ₹", round(average, 2))

max_gain = prices[1] - prices[0]
gain_day = 1

for i in range(1, len(prices)):
    gain = prices[i] - prices[i - 1]

    if gain > max_gain:
        max_gain = gain
        gain_day = i

print("\nBiggest Single-Day Gain = ₹", max_gain)
print("Between", days[gain_day - 1], "and", days[gain_day])
volatility = 0

for i in range(1, len(prices)):
    volatility += abs(prices[i] - prices[i - 1])

print("\nTotal Volatility =", volatility)
````

#Q45.

letter forward by N positions in the alphabet (wrapping around, Z + 1 = A). Preserve case. Keep digits,
spaces, and special characters UNCHANGED. Example: "Hello, India!" with shift 3 → "Khoor, Lqgld!"

````python
msg = input()
shift = int(input())

res = ""

for ch in msg:
    if 'a' <= ch <= 'z':
        res += chr((ord(ch) - 97 + shift) % 26 + 97)

    elif 'A' <= ch <= 'Z':
        res += chr((ord(ch) - 65 + shift) % 26 + 65)

    else:
        res += ch

print(res)
````

#Q46.
Write a function `calculate_gst(amount, gst_rate=18, gst_inclusive=False)`:
- If gst_inclusive=False: amount is base, add GST on top
- If gst_inclusive=True: amount already includes GST, extract it
- Return a tuple: (base_amount, gst_amount, total_amount)
- All values rounded to 2 decimals
Test cases:
 calculate_gst(1000) → (1000.00, 180.00, 1180.00)
 calculate_gst(1180, gst_inclusive=True) → (1000.00, 180.00, 1180.00)
 calculate_gst(500, gst_rate=5) → (500.00, 25.00, 525.00)

````python
def calculate_gst(amount, gst_rate=18, gst_inclusive=False):

    if gst_inclusive:
        base = amount / (1 + gst_rate/100)
        gst = amount - base
        total = amount
    else:
        base = amount
        gst = amount * gst_rate/100
        total = base + gst

    return (round(base,2), round(gst,2), round(total,2))


print(calculate_gst(1000))
print(calculate_gst(1180, gst_inclusive=True))
print(calculate_gst(500, gst_rate=5))
````

#Q47.
Write a RECURSIVE function `power(base, exp)` that computes base raised to exp, WITHOUT
using the `**` operator and WITHOUT using `pow()`. Handle exp = 0 (return 1) and negative exp
(return 1/power(base, -exp)). Test with: power(2, 10), power(5, 0), power(3, -2).

````python
def power(base, exp):

    if exp == 0:
        return 1

    if exp < 0:
        return 1 / power(base, -exp)

    return base * power(base, exp - 1)


print(power(2, 10))
print(power(5, 0))
print(power(3, -2))
````
